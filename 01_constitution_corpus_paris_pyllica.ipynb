# Ensemble 1 : https://gallica.bnf.fr/ark:/12148/cb32695181v/date OK

```python
# contenu fichier pyllica:
# jpgpress(url="https://gallica.bnf.fr/ark:/12148/cb32695181v/date", title="journal", year=xxxx, month=1, day=x, item=50, rate=7, firstpage=1, lastpage=20)
# extraction de 50 numéros par an, soit environ 1 numéro par semaine

# cas non traité: lorsque les périodes ont été numérisés en une seule fois, et apparaissent donc tous datés au premier jour de l'année 
# années concernées: 1752,1753,1754,1755,1781,1782
# exemple en 1751: un numéro daté du 1er janvier, de 800 pages, qui comprend en réalité toutes les parutions de l'année mises à la suite les unes des autres

# créer un nouveau dossier Annonces_Paris_année pour chaque année de parution disponible sur Gallica
# Dictionnaire: {année:jour du premier numéro disponible}

années_disponibles={"1756":"1", "1757":"6", "1758":"5", "1759":"5", "1760":"3", "1761":"1", "1762":"7", "1763":"6", "1764":"5", "1765":"3", "1773":"7", "1774":"6", "1775":"5", "1776":"4", "1778":"1"}
path_actionpyllica="./actionpyllicalabsjpgpress.py"
path_pyllica="./pyllicalabsjpgpress.py"

for année in années_disponibles:
    path="./Annonces_Paris_"+année
    os.mkdir(path)
    
# copier les deux fichiers pyllica dans chaque dossier
    shutil.copyfile(path_actionpyllica, path+"/actionpyllicalabsjpgpress.py")
    shutil.copyfile(path_pyllica, path+"/pyllicalabsjpgpress.py")
    
# modifier le contenu de actionpyllicalabsjpgpress "year" et "day"
    os.chdir(path)
    année_pyllica=re.compile("year=\d{4}")
    jour_pyllica=re.compile("day=\d")
    with open("actionpyllicalabsjpgpress.py", "r") as actionpyllica: 
        contenu=actionpyllica.read()
        contenu_avec_année=re.sub(année_pyllica, "year="+année, contenu)
        contenu_avec_jour=re.sub(jour_pyllica, "day="+années_disponibles[année], contenu_avec_année)
        
    with open("actionpyllicalabsjpgpress.py", "w") as actionpyllica_avec_date:
        actionpyllica_avec_date.write(contenu_avec_jour)
        
# lancer actionpyllicalabsjpgpress
    
    %run -i -t actionpyllicalabsjpgpress
```
