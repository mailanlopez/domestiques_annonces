```python
nlp = spacy.load("fr_core_news_md")
```

```python
def extraire_dix_premiers_mots(annonce):
    mots = annonce.split(' ')
    return ' '.join(mots[:10])
```

```python
df['Premiers_Mots'] = df['Annonces'].apply(extraire_dix_premiers_mots)
```

```python
def extraire_qualites(annonce):
    qualites=[]
    
    doc = nlp(annonce)
    for token in doc:
        if token.pos_=="ADJ":
            qualites.append(token.text)
        
    regex=", ?((?:d'|de |dont |au |aux )\w+'? ?\w+ (?:\w+)? (?:\w+)?),"
    qualites_regex=re.findall(regex, annonce, re.UNICODE)
    for i in qualites_regex:
        qualites.append(i)
        
    qual_a_retirer=["âgé", "âgée", "agé", "agée", "âge", "âgee", "age", "agee"]

    return " ".join(i for i in qualites if i not in qual_a_retirer)
```

```python
df['Qualités'] = df['Premiers_Mots'].apply(extraire_qualites)
```
