```python
from bertopic import BERTopic
import pandas as pd
```

```python
# corpus: uniquement les textes du fichier csv
df = pd.read_csv("./df_final.csv")
data = list(df['Annonces'])
```

```python
from spacy.lang.fr.stop_words import STOP_WORDS as fr_stop
french_stop= list(fr_stop)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer_model = CountVectorizer(stop_words=french_stop)
topic_model = BERTopic(language = "multilingual",vectorizer_model=vectorizer_model, min_topic_size=15)
topics, probs = topic_model.fit_transform(data)
```

```python
# sauvegarder le modèle


embedding_model = "sentence-transformers/all-MiniLM-L6-v2"
topic_model.save("/content/model_bert", serialization="safetensors", save_ctfidf=True, save_embedding_model=embedding_model)
```

```python
# PAR VILLE

# nouveau corpus équilibré

df_villes = df.groupby("Ville").sample(n=1000, random_state=1, replace=True)
data = list(df_villes['Annonces'])
```

```python
topic_model = BERTopic(language = "multilingual",vectorizer_model=vectorizer_model, min_topic_size=15)
topics, probs = topic_model.fit_transform(data)
```

```python
# par ville

classes = list(df['Ville'])

topics_per_class = topic_model.topics_per_class(data, classes=classes)
```

```python
topic_model.visualize_topics_per_class(topics_per_class, topics=[0, 1, 2, 6, 9, 12, 13, 14, 23])
```

```python
# DTM (dynamic topic modeling) avec BERTopic (topicsovertime)
# https://maartengr.github.io/BERTopic/getting_started/topicsovertime/topicsovertime.html#example

# nouveau corpus équilibré
df_année = df.groupby("Année").sample(n=100, random_state=1, replace=True)
data = list(df_année['Annonces'])
```

```python
topic_model = BERTopic(language = "multilingual",vectorizer_model=vectorizer_model, min_topic_size=15)
topics, probs = topic_model.fit_transform(data)
```

```python
# besoin de créer un dataset timestamps (avec les dates correspondant aux annonces)
# on crée une nouvelle colonne dans le dataframe: date (qui agrège année, jour, mois)

df['Date'] = pd.to_datetime(dict(year=df.Année, month=df.Mois, day=df.Jour))
```

```python
timestamps = list(df['Date'])
```

```python
topics_over_time = topic_model.topics_over_time(data, timestamps, nr_bins=20)
```

```python
topic_model.visualize_topics_over_time(topics_over_time, topics=[0, 2, 6, 7, 9, 10, 11, 13, 18, 20, 22, 23])
```
